```python
import pandas as pd
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
houseCostInitial = 120000
houseCostInflation = 0.05
periodInYears = 5
months = 12

houseValueFuture = -npf.fv(houseCostInflation, periodInYears, 0, houseCostInitial)
houseValueInPeriods = -npf.fv(houseCostInflation, np.arange(periodInYears + 1), 0, houseCostInitial)

print("House future value:")
print(houseValueFuture)

print("House value by year:")
print(houseValueInPeriods)
```

```python
bankInterestRate = 0.12
fixedInterestRate = bankInterestRate / months
periodsNumber = periodInYears * months
periodsArray = np.arange(1, periodsNumber + 1)

paymentBasicArray = -npf.ppmt(fixedInterestRate, periodsArray, periodsNumber, 0, houseValueFuture)
monthlyPayment = -npf.pmt(fixedInterestRate, periodsNumber, 0, houseValueFuture)

compoundInterestSum = np.cumsum(paymentBasicArray)
totalAccumulatedValue = np.sum(paymentBasicArray)

print("Save value at the end:")
print(totalAccumulatedValue)

print('Payment No | Account Balance | Payment')
dataFormatting = '{0:2d} {1:12.2f} {2:12.2f}'
for payment in periodsArray:
    print(dataFormatting.format(payment, compoundInterestSum[payment - 1], monthlyPayment))
```

```python
plt.plot(np.arange(periodInYears + 1), houseValueInPeriods, label='House Price Over Time')
plt.plot(periodsArray / months, compoundInterestSum, label='Bank Payments')
plt.legend()
plt.xlabel('Time (Years)')
plt.ylabel('Value in Currency')
plt.title('House Price and Bank Payments Over Time')
plt.show()
```
